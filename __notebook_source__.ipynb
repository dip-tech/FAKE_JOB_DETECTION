# IMPLEMENTING NATURAL LANGUAGE PROCESSING

```python
# IMPORTING RE
import re
# IMPORTING NLTK
import nltk
# DOWNLOAD STOPWORDS
nltk.download('stopwords')
# IMPORT STOPWORDS
from nltk.corpus import stopwords
# IMPORT PORTERSTEMMER
from nltk.stem.porter import PorterStemmer
```

```python
# CREATING CORPUS
corpus=[]
for indx in range(0,len(dataset['description'])):
    temp=re.sub('[^A-Za-z]',' ',dataset['description'][indx])
#     print(indx)
    temp=temp.lower()
    temp=re.sub(' +',' ',temp)
    temp=temp.split(' ')
    PS=PorterStemmer()
    sWords=stopwords.words('english')
    temp=[PS.stem(w) for w in temp if w not in set(sWords)]
    temp=' '.join(temp)
    temp=temp.strip()
    corpus.append(temp)
```

```python
dataset['description']
```

```python
# DISPLAY CORPUS
corpus
```

```python
# DISPLAY DATASET
dataset.head(5)
```

```python
# CREATING BAG OF WORDS

# IMPORTING COUNT VECTORIZER
from sklearn.feature_extraction.text import CountVectorizer
# CREATING OBJECT OF COUNT VECTORIZER
CV=CountVectorizer(max_features=5000)
# CREATING INDEPENDENT VARIABLE
X=CV.fit_transform(corpus).toarray()
```

```python
# DISPLAY X
X
```

```python
# GET FEATURE NAME FROM CV
CV.get_feature_names()
```

```python
# ADDING HAS COMPANY LOGO IN X
X2=dataset.loc[:,'has_company_logo'].values
# DISPLAYING X2
X2
```

```python
# CONVERTING X2 INTO 2D ARRAY
X2=X2.reshape(len(X2),1)
# DISPLAYING X2
X2
```

```python
# NOW X2 CONCAT WITH INDEPENDENT VARIABLE X
X=np.concatenate((X,X2),axis=1)
```

```python
# DISPLAY X
X
```

```python
# CREATING DEPENDENT VARIABLE
Y=dataset.iloc[:, -1].values
```

```python
# DISPLAY Y
len(Y)
```

# DATA PREPROCESSING

```python
# IMPORTING TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
# SPLITTING DATASET INTO TRAIN SET AND TEST SET
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=0,stratify=Y)
```

```python
# DISPLAY Y TRAIN
Y_train
```

```python
# DISPLAY Y TEST
Y_test
```

```python
# DISPLAY X TRAIN
X_train
```

```python
# DISPLAY X TEST
X_test
```

# NAIVE BAYES MODEL

```python
# IMPORTING NAIVE BAYES MODEL
from sklearn.naive_bayes import GaussianNB
# CREATING MODEL
NB=GaussianNB()
# TRAIN MODEL USING TRAINING SET
NB.fit(X_train,Y_train)
```

```python
# PREDICTING RESULT
Y_NB=NB.predict(X_test)
```

```python
# COMPARE RESULT
RESULT=pd.DataFrame({
    'Y_NB':Y_NB,
    'Y_TRUE':Y_test
})
RESULT
```

```python
# CALCULATING ACCURACY AND CONFUSION MATRIX
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
# ACCURACY SCORE
accuracy_score(Y_test,Y_NB)*100
```

```python
# CONFUSION MATRIX
confusion_matrix(Y_test,Y_NB)
```

```python
# CLASSIFICATION REPORT
print(classification_report(Y_test,Y_NB))
```
